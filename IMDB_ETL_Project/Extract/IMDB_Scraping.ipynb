```python
from selenium import webdriver
import pandas as pd
```

```python
driver = webdriver.Chrome()
```

```python
driver.get("https://www.imdb.com/chart/top/")
```

```python
movies_names = driver.find_elements("xpath","//h3[@class='ipc-title__text ipc-title__text--reduced']")
```

```python
for movie in movies_names:
    print(movie.text)
```

```python
movies_card = driver.find_elements("xpath","//div[@class='sc-ec40e84d-0 dTHKNo']")
movies_data =[]

for movie in movies_card :
    movie_Name = movie.find_element("xpath",".//h3[@class='ipc-title__text ipc-title__text--reduced']").text
    details = movie.find_elements("xpath",".//span[@class='sc-15ac7568-7 cCsint cli-title-metadata-item']")
    movie_year = details[0].text if len(details) > 0 else "NULL"
    movie_duration = details[1].text if len(details) > 1 else "NULL"
    movie_certificate = details[2].text if len(details) > 2 else "NULL"

    movie_Rate = movie.find_element("xpath",".//span[@class='ipc-rating-star--rating']").text
    
    try:
        movie_URL = movie.find_element("xpath",".//a[@class='ipc-lockup-overlay ipc-focusable']").get_attribute("href")
    except:
        movie_URL = "NULL"

    
    movies_data.append({
        "movie Name": movie_Name , 
        "movie year": movie_year , 
        "movie Rate" : movie_Rate ,
        "movie duration" : movie_duration ,
        "movie certificate" : movie_certificate ,
        "movie URL" : movie_URL
        })
    
movies_df = pd.DataFrame(movies_data)

movies_df.to_csv("IMDB_top_250_movies.csv" , index= False ,encoding="utf-8")

print(movies_df.head())
```
